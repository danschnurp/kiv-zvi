```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread("C:/Users/dartix/PycharmProjects/kiv-zvi/images/photographer.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')

'''
Prepare simple kernels for differences in x and y directions
'''
kernelX = np.ones((2,1))
kernelX[0,0] = -1
kernelY = np.ones((1,2))
kernelY[0,0] = -1

'''
Apply the filters by convolutional operator
'''
edgesX = cv2.filter2D(img, -1, kernelX)
edgesY = cv2.filter2D(img, -1, kernelY)
plt.imshow(edgesY, cmap='gray')
```

```python
'''
Apply Scharr filter
'''
mask =np.array([[3, 0,-3],
                [10, 0, -10],
                [3,0,-3]])
edgesX = cv2.filter2D(img, -1, mask)
edgesY = cv2.filter2D(img, -1, mask.T)

plt.imshow(edgesX, cmap='gray')
plt.show()
plt.imshow(edgesY, cmap='gray')
```
