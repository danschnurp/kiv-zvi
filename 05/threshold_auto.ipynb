```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("photographer.png", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([img], [0], None, [256], [0,255])
hist = np.reshape(hist, (hist.shape[0],))
p = hist / np.sum(hist)
plt.plot(p, color = 'blue')
plt.xlim([0,256])
plt.show()
```
